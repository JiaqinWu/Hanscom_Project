## Web Scrape by Using API
### Website: https://www.flightera.net/en/airport/Bedford/KBED/departure/2023-06-01%20%2000_00?

```python
import pandas as pd
import requests
```

```python
url = "https://flightera-flight-data.p.rapidapi.com/airport/flights"

querystring = {"direction":"departure","time":"2023-06-29T13:05:00Z","ident":"KBED"}

headers = {
	"X-RapidAPI-Key": "93a1c2e6bemsh1c89f0bd4d8f7a1p1c4068jsn985518990be9",
	"X-RapidAPI-Host": "flightera-flight-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

# print(response.json())
data1 = response.json()
# Create a dataframe from the JSON data
df1 = pd.DataFrame(data1)
# Print the dataframe
df1[['flnr', 'date']] = df1['flights'].apply(lambda x: pd.Series([x['flnr'], x['date']]))
df1
```

```python
# Define the API endpoint and headers
url = "https://flightera-flight-data.p.rapidapi.com/flight/info"
headers = {
    "X-RapidAPI-Key": "93a1c2e6bemsh1c89f0bd4d8f7a1p1c4068jsn985518990be9",
    "X-RapidAPI-Host": "flightera-flight-data.p.rapidapi.com"
}

# Create an empty dataframe to store the API responses
df_fina = pd.DataFrame()

# Iterate over the rows of df1 and make API calls
for index, row in df1.iterrows():
    # Get the flight details from df1
    flnr = row['flnr']
    date = row['date']
    
    # Set the querystring parameters
    querystring = {"flnr": flnr, "date": date}
    
    # Make the API request
    response = requests.get(url, headers=headers, params=querystring)
    
    # Check if the API request was successful
    if response.status_code == 200:
        # Get the JSON data from the API response
        data = response.json()
        
        # Create a dataframe from the JSON data
        df_temp = pd.DataFrame(data)
        
        # Concatenate the dataframe to the final dataframe
        df_fina = pd.concat([df_fina, df_temp], ignore_index=True)
    else:
        # Handle the case when the API request fails
        print(f"API request for flight {flnr} on {date} failed")

# Print the resulting DataFrame
df_fina
```
